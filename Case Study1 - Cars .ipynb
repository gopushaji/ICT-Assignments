# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
male_buyers=len(cars_data[cars_data['Buyer Gender']=='Male'].index)
female_buyers=len(cars_data[cars_data['Buyer Gender']=='Male'].index)
print('Male Buyers: ', male_buyers, '\tFemale Buyers: ', female_buyers)
```

```python
plt.hist(cars_data['Buyer Gender'])
plt.title('Distribution on Gender',fontsize=15)
plt.xlabel('Gender Classification')
plt.ylabel('Number')
```

# 4. Find the top 5 cars based on their sales price.

```python
sale_price=cars_data['Sale Price']
sale_price_sorted=sale_price.sort_values(ascending=False)
top_index=sale_price_sorted[0:5].index
pd.DataFrame({'Make':list(cars_data['Make'][top_index]),
              'Model':list(cars_data['Model'][top_index]),
              'Sale Price':list(cars_data['Sale Price'][top_index])},index=[1,2,3,4,5])
```

# 5. Find the least 5 cars based on their Resell price.

```python
resell_price=cars_data['Resell Price']
resell_price_sorted=resell_price.sort_values(ascending=True)
least_index=resell_price_sorted[0:5].index
pd.DataFrame({'Make':list(cars_data['Make'][least_index]),
              'Model':list(cars_data['Model'][least_index]),
              'Resell Price':list(cars_data['Resell Price'][least_index])},index=[1,2,3,4,5])
```
