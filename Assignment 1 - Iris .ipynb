```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read the dataset to python environment.

```python
iris_data= pd.read_excel('iris.xls')
```

```python
iris_data.columns
```

```python
np.mean(iris_data)
```

```python
iris_data.isnull().values.any()
```

# 5. Perform meaningful visualizations using the dataset

```python
plt.hist(iris_data['Classification'])
plt.title('Distribution of Each Variety')
plt.xlabel('Classification')
plt.ylabel('Number')
```

```python
sns.pairplot(iris_data[['SL', 'SW', 'PL', 'PW']])
```

```python
plt.figure(figsize=(14,10))

plt.subplot(2,2,1)
plt.title('SW vs SL')
sns.scatterplot(hue='Classification',y='SW',x='SL',data=iris_data)

plt.subplot(2,2,2)
plt.title('PW vs PL')
sns.scatterplot(hue='Classification',y='PW',x='PL',data=iris_data)

plt.subplot(2,2,3)
plt.title('SL vs PL')
sns.scatterplot(hue='Classification',y='SL',x='PL',data=iris_data)

plt.subplot(2,2,4)
plt.title('SW vs PW')
sns.scatterplot(hue='Classification',y='SW',x='PW',data=iris_data)
```

```python
plt.figure(figsize=(14,10))

plt.subplot(2,2,1)
plt.title('SL vs Classification')
sns.swarmplot(x='Classification',y='SL',data=iris_data,s=2.5)

plt.subplot(2,2,2)
plt.title('SW vs Classification')
sns.swarmplot(x='Classification',y='SW',data=iris_data,s=2.5)

plt.subplot(2,2,3)
plt.title('PL vs Classification')
sns.boxplot(x='Classification',y='PL',data=iris_data)

plt.subplot(2,2,4)
plt.title('PW vs Classification')
sns.boxplot(x='Classification',y='PW',data=iris_data)
```
